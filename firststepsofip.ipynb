```
import numpy as np
from matplotlib import pyplot as plt
from skimage import data, color
from skimage.filters import try_all_threshold
from skimage.filters import threshold_otsu, threshold_yen
```

# **Show Images**

```
def show_image(image, title='Image', cmap_type = 'gray') :
  plt.imshow(image, cmap = cmap_type)
  plt.axis('off')
  plt.title(title)
  plt.show()
```

A simple function to show our images.

```
retina = data.retina()
```

Load the RGB-3 image from skimage library.

```
gray_scaled_retina = color.rgb2gray(retina)
```

Convert the image to grayscale.

```
show_image(retina, 'Original RGB image')

show_image(gray_scaled_retina, 'Grayscale image')
```

Show the RGB-3 and grayscaled versions of the original image.

# **Numpy Images**

```
view = plt.imread("view.jpg")
type(view)
```

Read an image with matplatlib.pyplot as an ndarray.

```
plt.imshow(view)
plt.axis('off')
plt.show()
```

```
red = view[:, :, 0]
green = view[:, :, 1]
blue = view[:, :, 2]
```

Obtaining the red, green and blue values of the image.

```
plt.subplot(1, 3, 1)
plt.imshow(red,'gray')


plt.subplot(1, 3, 2)
plt.imshow(green,'gray')

plt.subplot(1, 3, 3)
plt.imshow(blue,'gray')

plt.axis('off')
plt.tight_layout()
plt.show()
```

Different intensities of each color is showed.

```
plt.subplot(1, 3, 1)
plt.imshow(view)
plt.title('original')
plt.axis('off')

plt.subplot(1, 3, 2)
vertically_flipped_view = np.flipud(view)
plt.imshow(vertically_flipped_view)
plt.title('vertically flipped')
plt.axis('off')

plt.subplot(1, 3, 3)
completely_flipped_view = np.fliplr(vertically_flipped_view)
plt.imshow(completely_flipped_view)
plt.title('completely flipped')
plt.axis('off')

plt.tight_layout()
plt.show()
```

Flipped versions of the view has shown.

```
plt.hist(red.ravel(), bins = 256)

plt.title('Red Histogram')
plt.show()
```

A histogram to display the 256 different intensities of the red color.

# **Thresholding**

Isolation of objects (object detection, face detection)

* Global Thresholding : for uniform background
* Local Thresholding : for uneven background


```
gradient = plt.imread("gradient.jpg")

gradient = color.rgb2gray(gradient)

fig, ax = try_all_threshold(gradient, figsize=(10, 8), verbose=False)
plt.show()
```

Try all the global threshold types and pick one of them.


```
global_jpg = plt.imread("global.jpg")
global_jpg = color.rgb2gray(global_jpg)

threshold_for_yen = threshold_yen(global_jpg)
binary_global = global_jpg > threshold_for_yen

plt.subplot(1, 3, 1)
plt.imshow(global_jpg)
plt.title("Grayscaled image")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(binary_global)
plt.title("Yen Threshold")
plt.axis('off')

plt.tight_layout()
plt.show()
```

An implementation of yen thresholding.
